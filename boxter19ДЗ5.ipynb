```python
def surn(row):
    if type(row['Name'])!=str and type(row['Name'])!=unicode:
        return 'Weird name'
    else:
        return row['Name'].split(',')[0]
```

```python
df['Surname']=df.apply(surn, axis=1)
df.sample(20)
```
